```python
clickbaitdata = list(zip(data,datatruth))
#print (clickbaitdata[:2])
```

```python
# counting numbers of clickbait and nonclickbait

countCB = 0
countNCB = 0

for instance in clickbaitdata:
  answer = (instance[1]['truthClass'])
  if answer == "clickbait":
    countCB+=1
  elif answer == 'no-clickbait':
    countNCB +=1
    
print (countCB)
print (countNCB)
```

```python
countCaptionsForClickbait = 0
countCaptionsForNonClickbait = 0

for instance in clickbaitdata:
  answer = (instance[1]['truthClass'])
  if answer =="no-clickbait":
    test = instance[0]['targetCaptions']
    num = len(test)
    countCaptionsForNonClickbait +=num
  if answer =="clickbait":
    test = instance[0]['targetCaptions']
    num = len(test)
    countCaptionsForClickbait +=num
print (countCaptionsForClickbait)
print (countCaptionsForNonClickbait)

averageNumberOfCaptionsForClickbait = countCaptionsForClickbait / countCB
averageNumberOfCaptionsForNonClickbait = countCaptionsForNonClickbait / countNCB

print(averageNumberOfCaptionsForClickbait)
print(averageNumberOfCaptionsForNonClickbait)

print("conclusion: NCB has a significantly higher number of target captions than CB")
```

```python
# percentage of CN and NCB with a caption length of less than 5 words or more than 25 words, these two numbers are picked manually, if its too short then its not informative

countNumberOfCaptionsCB = 0      
countNumberOfCaptionsNonCB = 0    
NumberOfCaptionsLessThan5wordsCB = 0
NumberOfCaptionsLessThan5wordsNonCB = 0

for instance in clickbaitdata:
  answer = (instance[1]['truthClass'])
  if answer =="clickbait":
    targetCaptions = instance[0]['targetCaptions']
    for captions in targetCaptions:
      countNumberOfCaptionsCB+=1
      lengthOfThisCaption = len(captions.split())
      if lengthOfThisCaption <5:
        NumberOfCaptionsLessThan5wordsCB +=1
  if answer == "no-clickbait":    
    targetCaptions = instance[0]['targetCaptions']
    for captions in targetCaptions:
      countNumberOfCaptionsNonCB+=1
      lengthOfThisCaption = len(captions.split())
      if lengthOfThisCaption <5:
        NumberOfCaptionsLessThan5wordsNonCB +=1
           
#print(countNumberOfCaptionsCB)
#print(countNumberOfCaptionsNonCB)  
#print(NumberOfCaptionsLessThan5wordsCB)
#print(NumberOfCaptionsLessThan5wordsNonCB)  

averageCB = NumberOfCaptionsLessThan5wordsCB/countNumberOfCaptionsCB
averageNonCB = NumberOfCaptionsLessThan5wordsNonCB/countNumberOfCaptionsNonCB

print("percentage of clickbait captions with less than 5 words (4 and below) is : " + str(averageCB *100) + "%")
print("percentage of Non-clickbait captions with less than 5 words (4 and below) is : " + str(averageNonCB *100) + "%")

# less than 5 is 33.2% CB and 26.7% NCB
# less than 4 is 28.3% CB and 20.9% NCB
# less than 3 is 22.3% CB and 15.4% NCB
# less than 2 is 13.3% CB and 8.8% NCB

            
```

```python
#comparing length of more than 15,20,25,30 words in each caption now
countNumberOfCaptionsCB = 0      
countNumberOfCaptionsNonCB = 0    
NumberOfCaptionsMoreThan25wordsCB = 0
NumberOfCaptionsMoreThan25wordsNonCB = 0

for instance in clickbaitdata:
  answer = (instance[1]['truthClass'])
  if answer =="clickbait":
    targetCaptions = instance[0]['targetCaptions']
    for captions in targetCaptions:
      countNumberOfCaptionsCB+=1
      lengthOfThisCaption = len(captions.split())
      if lengthOfThisCaption >30:
        NumberOfCaptionsMoreThan25wordsCB +=1
  if answer == "no-clickbait":    
    targetCaptions = instance[0]['targetCaptions']
    for captions in targetCaptions:
      countNumberOfCaptionsNonCB+=1
      lengthOfThisCaption = len(captions.split())
      if lengthOfThisCaption >30:
        NumberOfCaptionsMoreThan25wordsNonCB +=1
           
print(countNumberOfCaptionsCB)
print(countNumberOfCaptionsNonCB)  
print(NumberOfCaptionsMoreThan25wordsCB)
print(NumberOfCaptionsMoreThan25wordsNonCB)  

averageCB = NumberOfCaptionsMoreThan25wordsCB/countNumberOfCaptionsCB
averageNonCB = NumberOfCaptionsMoreThan25wordsNonCB/countNumberOfCaptionsNonCB

print("percentage of clickbait captions with more than 25 words is : " + str(averageCB *100) + "%")
print("percentage of Non-clickbait captions with more than 25 words is : " + str(averageNonCB *100) + "%")

#more than 15 is 22.8% CB and 37.1% NCB
#more than 20 is 16.3% CB and 29.3% NCB
#more than 25 is 12.5% CB and 21.8% NCB
#more than 30 is 9.9% CB and 16.3% NCB
```

```python
#count the number of paragraphs on average for CB and nonCB
counterParagraphCB = 0
counterParagraphNonCB = 0

for instance in clickbaitdata:
  answer = (instance[1]['truthClass'])
  if answer =="clickbait":
    targetParagraphs = instance[0]['targetParagraphs']
    lengthP = len(targetParagraphs)
    counterParagraphCB += lengthP
  if answer =="no-clickbait":
    targetParagraphs = instance[0]['targetParagraphs']
    lengthP = len(targetParagraphs)
    counterParagraphNonCB += lengthP    
    
    
averageParagraphsForCB = counterParagraphCB / countCB
averageParagraphsForNonCB = counterParagraphNonCB / countNCB

print("average number of paragraphs in Clickbait is " + str(averageParagraphsForCB))
print("average number of paragraphs in Non-Clickbait is " + str(averageParagraphsForNonCB))
```
